```python
# 판다스 예제
import pandas as pd
```

```python
# 리스트를 이용 => Series
list = [100, 20, 30, 40]
scores = pd.Series(list, index=["국어", "영어", "수학", "과학"])

print(scores)

# 딕셔너리를 이용 => DataFrame
scores_dic = {
	'과목' : ["국어", "영어", "수학", "과학"],
	'성적' : [100, 90, 80, 70],
	'출석' : [100, 100, 0, 80],
	'수행' : [90, 100, 90, 30]
}

df = pd.DataFrame(scores_dic)
print(df)
```

```python
# 데이터 탐색 예제
print(f"상위 2열 : \n{df.head(2)}")
print(f"하위 2열 : \n{df.tail(2)}")
print(f"info : \n{df.info()}")
print(f"describe : \n{df.describe()}")
print(f"shape : \n{df.shape}")
```

```python
def bar():
  print("-"*30)
```

```python

# 데이터 선택 및 필터링
print("데이터 프레임")
bar()
print(df)
bar()

# 과목들
print(df['과목'])
bar()

# 국어 성적 성택
print(f"국어 성적 : \n{df.loc[0]}")
bar()

# 과목의 수행평가 점수 조회
print("과목과 수행 평가 조회")
# 조회하려는 열이 여러개인 경우 열 목록을 리스트로 전달
print(df[['과목','수행']])
bar()
```

```python
# 데이터 결측치 처리
df2 = pd.DataFrame({
	'이름' : ['홍길동', '이순신', '고길동'],
 	'나이' : [21, None, 40],
	'주소' : ['조선', '조선', '서울'],
 	'가족' : [None, '있음','있음']
})
```

```python
# 시험 열 추가
df['시험'] = df['성적']
print(df)
```

```python
# 성적 열 삭제
df.drop('성적', axis=1, inplace=True)
print(df)
```

```python
# 총점 열 추가
df['총점'] = df['출석'] + df['수행'] + df['시험']
print(df)
```

```python
# 비고 열 추가 및 값 추가
df['비고'] = '없음'
print(df)
```

```python
# 0번지 행 삭제
print(df.drop(0))
```

```python
# 과목명이 과학인 행 삭제
index = df[df['과목'] == '과학'].index
print(df.drop(index))
```

```python
# 시험 성적 순으로 정렬 : 오름차순
print(df.sort_values(by='시험'))
```

```python
# 시험 성적 순으로 정렬 : 내림차순
print(df.sort_values(by='시험', ascending=False))
```

```python
# 데이터 결함 및 그룹 예제
df3 = pd.DataFrame({
	"이름" : ["홍길동", "고길동", "둘리"],
 	"국어성적" : [100, 90, 80]
})
df4 = pd.DataFrame({
	"이름" : ["홍길동", "고길동", "둘리"],
	 "영어성적" : [80, 70, 80]
})
```

```python
# 국어 성적과 영어 성적 합치기
total = pd.merge(df3, df4, on="이름")
print(total)
```

```python
total2 = pd.DataFrame({
  "이름" : ["개똥이", "하니", "또치"],
  "국어성적" : [50, 50, 90],
  "영어성적" : [30, 40, 90]
})

total = pd.concat([total, total2])
print(total)
```

```python
# 반 열을 추가
total['반'] = 1
print(total)
```

```python
total.loc[total['이름'].str.len() < 3,'반'] = 2
print(total)
```

```python
# 반 별 국어 성적 평균
print(total.groupby("반")["국어성적"].mean())
# 반 별 국어 성적 평균, 총합
print(total.groupby("반")["국어성적"].agg(['mean','sum']))
```

```python
# 반별로 정렬
total = total.sort_values(by = '반')

# index 초기화
total = total.reset_index(drop=True) # drop=True는 기존 인덱스를 삭제
print(total)
```

```python
# 중복 데이터를 마지막에 추가
total.loc[len(total)] = ["또치", 90, 90, 2]

print(total.drop_duplicates())
```
