```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# 네이버 검색 결과화면에서 뉴스 제목과 링크 추출하기

keyword = 'ai'

url = f'https://search.naver.com/search.naver?ssc=tab.news.all&where=news&sm=tab_jum&query={keyword}'

# request로 url로 요청해서 응답을 가져옴
response = requests.get(url)
```

```python
# 상태 코드 확인 후 200(OK)일 때 소스코드 확인
if response.status_code == 200:
    print("성공적으로 연결되었습니다!")
    # HTML 소스코드 확인
    # print(response.text[:500])
    
    # 가져온 결과에서 다음을 만족하는 요소들을 가져옴
    # .sds-comps-full-layout a[data-heatmap-target=".tit"]
    
    # # HTML 소스를 넘겨받아 BeautifulSoup 객체 생성
    soup = BeautifulSoup(response.text, 'lxml')
    
    # 가져온 요소들을 콘솔에 출력
    links = soup.select('.sds-comps-full-layout a[data-heatmap-target=".tit"]')
    
    titles = []
    hrefs = []
    
    for link in links:
      # 제목과 링크를 추출해서 콘솔에 출력
      title = link.text
      href = link.get('href')
      
      # 제목과 링크를 제목 리스트, 링크 리스트에 각각 뒤에 추가
      titles.append(title)
      hrefs.append(href)
      
      # 판다스를 이용하여 DataFrame을 생성
      # 추가할 때 열 이름은 title과 href로 지정
      data = {
        'title': titles,
        'href': hrefs
        }
      df = pd.DataFrame(data)
      
      # csv 파일로 저장
      # # index=False를 안 하면 0, 1, 2... 같은 인덱스 번호도 같이 저장돼서 보통 False를 써.
      df.to_csv(f'naver_nl_{keyword}.csv', index=False, encoding='utf-8-sig')
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time

keyword = input("키워드를 입력하세요 : ")

print(f"입력하신 키워드는 : {keyword} 입니다.")

def search(keyword, start:int=1):

  titles = []
  hrefs = []

  for i in range(1, 51, 10):

    url = 'https://search.naver.com/search.naver'

    params = {
        'where': 'news',
        'query': keyword,
        'start': i
    }

    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)',
        'Referer':'https://www.naver.com/'
    }

    response = requests.get(url, params=params, headers=headers)

    if response.status_code == 200:
      print(f"{i} 페이지 접속 성공")

      soup = BeautifulSoup(response.text,'lxml')

      links_divs = soup.select('.n8_DDlCjHxLYRO50CY50')

      for link_div in links_divs:

        title_link = link_div.select_one('.XEtVZ4N7DI2Pv29xe7f3')
        href_link = link_div.select_one('.RQNKk8QZaQZble3gmgEj [data-heatmap-target=".nav"]')

        if title_link and href_link:
          titles.append(title_link.text)
          hrefs.append(href_link.get("href"))

    time.sleep(2)

  data = {
    'title': titles,
    'href': hrefs
  }

  df = pd.DataFrame(data)

  df.to_csv(f'naver_news_{keyword}.csv', index=False, encoding='utf-8-sig')

  print("뉴스 저장 완료")

search(keyword)
```
