```python
# 2차원 좌표 평면에 점을 나타나는 Point 클래스를 선언

# Point 클래스의 객체를 생성
class Point:
    # 클래스 변수
    name : str = "포인트 클래스"
    # 생성자
    def __init__(self, x:int = 0, y:int = 0):
        # 인스턴스 변수
        self.__x = x
        self.__y = y
    # 함수
    def print(self):
        print(f"({self.__x},{self.__y})")

    @property
    def x(self):
        return self.__x
    @x.setter
    def x(self, x):
        self.__x = x
    @property
    def y(self):
        return self.__y
    @x.setter
    def y(self, y):
        self.__y = y

print(Point.name)
p1 = Point(0,0)
p2 = Point(1,1)
p3 = Point()
p1.print()
p2.print()
p3.x = 100
p3.y = 100
p3.print()
```

```python
# 은행 계과 클래스
# 소유자(owner), 잔액(balance)
# 기능 : 출금(withdraw), 입금(deposit)

# Account 클래스의 객체를 생성
class Account:
     # 클래스 변수
    name : str = "계좌 잔액"
    
    # 생성자
    def __init__(self, owner, balance):
        self.__owner = owner
        self.__balance = balance

    # 함수
    def print(self):
        print(f"({self.__owner},{self.__balance})")

    @property
    def owner(self):
        return self.__owner
    @owner.setter
    def owner(self, owner):
        self.__owner = owner
    @property
    def balance(self):
        return self.__balance
    @balance.setter
    def balance(self, balance):
        self.__balance = balance
           
    def withdraw(self, withdraw):
        if self.__balance < withdraw:
            print("계좌 잔액이 부족합니다.")
            print(f"{self.__owner}님 계좌에서 출금 하려는 금액 : {withdraw:,}원")
        else:
            print(f"{self.__owner}님 출금전 계좌 잔액 : {self.__balance:,}원")
            self.__balance -= withdraw
            print(f"{self.__owner}님 계좌에서 출금하신 금액 : {withdraw:,}원")
            
    def deposit(self, deposit):
        print(f"{self.__owner}님 입금전 계좌 잔액 : {self.__balance:,}원")
        print(f"{self.__owner}님 계좌로 입금하신 금액 : {deposit:,}원")
        self.__balance += deposit

    def printInfo(self):
        print(f"{self.__owner}님 계좌 잔액 : {self.__balance:,}원")
        
# 홍길동이 계좌 개설
print("====================")
account = Account("홍길동", 0)
account.printInfo()

# 100000(10만) 입금
print("====================")
account.deposit(100000)
account.printInfo()

# 50000(5만) 출금
print("====================")
account.withdraw(50000)
account.printInfo()

# 70000(7만) 출금 => 잔액 부족 알림
print("====================")
account.withdraw(70000)
account.printInfo()
```

```python
# 도서 대여 시스템에서 사용할 도서 클래스 선언 및 활용
# 도서 클래스
# 저자(author), 도서명(title), 대출 가능 여부(is_borrowed)
# 기능 : 대출, 반납, 정보확인
class Book:
    # 생성자
    def __init__(self, title : str, author : str):
        self.__title : str = title
        self.__author : str = author
        self.__is_borrowed = False
        self.__borrower = None

    # 대출
    def borrow_book(self, user_name : str):
        print("====================")
        print(f"{user_name}님이 {self.__title}을/를 대출을 시도합니다.")
        if self.__is_borrowed:
            print(f"{self.__title}은/는 미반납 상태입니다.")
            # 이미 대출 중인 도서라 대출이 불가능합니다.
            print("이미 대출 중인 도서라 대출이 불가능합니다.")
            print(f"{user_name}님이 {self.__title}을/를 대출하지 못 하였습니다.")
        else:
            self.__is_borrowed = True
            self.__borrower = user_name
            # 회원1님이 파이썬 정석을 대출했습니다.
            print(f"{self.__title}은/는 반납 상태입니다.")
            print(f"{user_name}님이 {self.__title}을/를 대출했습니다.")
        
    # 반납
    def return_book(self, user_name : str):
        print("====================")
        if self.__is_borrowed and self.__borrower == user_name:
            # 회원1님이 파이썬 정석을 반납했습니다. 해당 도서는 대출이 가능합니다.
            print(f"{user_name}님이 {self.__title}을/를 반납했습니다. 해당 도서는 대출이 가능합니다.")
            self.__is_borrowed = False
            self.__borrower = None
        
    # 정보확인
    def show_info(self):
        print("====================")
        print(f"제목 : {self.__title}")
        print(f"저자 : {self.__author}")
        if self.__is_borrowed:
            print(f"대출 여부 : 대출 중")
        else:
            print(f"대출 여부 : 대출 가능")

book1 = Book("파이썬 정석", "홍길동")
book2 = Book("AI 원리", "고길동")

book1.borrow_book("회원1") # 회원1님이 파이썬 정석을 대출했습니다.
book1.borrow_book("회원2") # 이미 대출 중인 도서라 대출이 불가능합니다.
book1.return_book("회원1") # 회원1님이 파이썬 정서을 반납했습니다. 해당 도서는 대출이 가능합니다.

book1.show_info()
# 제목 : 파이썬 정석
# 저자 : 홍길동
# 대출 여부 : 대출 가능 # 이미 대출 중인 경우 대출 중
    
```
