```python
# 문장을 입력받아 각 단어가 몇번 사용 됐는지 판별하는 코드를 작성하세요
# 입력 : 안녕 내 이름 홍길동 네 이름 뭐니 내 이름 임꺽정
# 출력
# 안녕 : 1
# 내 : 2
# 이름 : 3
# 홍길동 : 1
# 네 : 1
# 뭐니 : 1
# 임꺽정 : 1

# 콘솔에서 문장 입력받기
sentence = input("문장을 입력하세요 : ")

def analyze_words(sentence):
    # 입력 받은 문장을 리스트에 넣기, 넣는 기준은 공백을 간격으로 분리해서 단어 구분
    words = sentence.split()

    dic = {}
    # 리스트에 들어간 요소들을 출력, 출력할때 리스트내에서 몇번 중복되는지 숫자로 표현
    for word in words :
        if word in dic :
            dic[word] += 1
        else:
            dic[word] = 1
    return dic

def analyze_words2(sentence):
    # 입력 받은 문장을 리스트에 넣기, 넣는 기준은 공백을 간격으로 분리해서 단어 구분
    words = sentence.split()

    dic = {}
    # 리스트에 들어간 요소들을 출력, 출력할때 리스트내에서 몇번 중복되는지 숫자로 표현
    for word in words :
        if word not in dic :
            dic[word] = words.count(word)
            
    return dic

def print_dic(dic):
    for key, value in dic.items():
        print(f"\"{key}\" : {value}")

print_dic(analyze_words(sentence))
print("===============")
print_dic(analyze_words2(sentence))
```
