```python
# Phone 클래스를 추가
# 모델명(model), 가격(price)
# 기능 : 생성자 - 모델명과 가격을 초기화, 모델조회(display)

# IPhone 클래스를 추가(Phone 클래스를 상속)
# 페이스아이디적용여부(face_id)
# 기능 : 생성자 - 모델명과 가격, 페이스아이디 적용 여부를 초기화,
# 			모델조회(display) - 모델명, 가격, 페이스적용여부

# Galaxy 클래스를 추가(Phone 클래스를 상속)
# 펜 지원(s_pen)
# 기능 : 생성자 - 모델명, 가격, 펜지원여부를 초기화
# 			모델조회(display) - 모델명, 가격, 펜지원 여부

# IPhone : IPhone15, 1250000, True
# IPhone : IPhoneSE, 6500000, False
# Galaxy : S24 Ultra, 1690000, True
# Galaxy : Z Flip 5, 1390000, False

# ==========================================================

# Phone 클래스를 추가
class Phone:

	# 모델명(model), 가격(price)
	# 기능 : 생성자 - 모델명과 가격을 초기화, 모델조회(display)
	def __init__(self, model : str, price :  int):
		self.__model : str = model
		self.__price :  int = price
			
	def display(self):
		print(f"모델명 : {self.__model}")
		print(f"가격 : {self.__price:,}원")

class IPhone(Phone):

	# 기능 : 생성자 - 모델명과 가격, 페이스아이디 적용 여부를 초기화,
	# 			모델조회(display) - 모델명, 가격, 페이스적용여부
	def __init__(self, model : str, price :  int, face_id : bool ):
		super().__init__(model, price)
		self.__model : str = model
		self.__price :  int = price
		self.__face_id : bool = face_id
	
	def display(self):
		print(f"모델명 : {self.__model}")
		print(f"가격 : {self.__price:,}원")
		print(f"페이스적용 여부 : {"O" if self.__face_id else "X"}")

class Galaxy(Phone):

	# 기능 : 생성자 - 모델명, 가격, 펜지원여부를 초기화
	# 			모델조회(display) - 모델명, 가격, 펜지원 여부
	def __init__(self, model : str, price :  int, s_pen : bool ):
		super().__init__(model, price)
		self.__model : str = model
		self.__price :  int = price
		self.__s_pen : bool = s_pen
	
	def display(self):
		print(f"모델명 : {self.__model}")
		print(f"가격 : {self.__price:,}원")
		print(f"펜지원 여부 : {"O" if self.__s_pen else "X"}")

# IPhone : IPhone15, 1250000, True
# IPhone : IPhoneSE, 6500000, False
# Galaxy : S24 Ultra, 1690000, True
# Galaxy : Z Flip 5, 1390000, False
print("=========================")
p1 = IPhone("IPhone15", 1250000, True)
p1.display()

print("=========================")
p2 = IPhone("IPhoneSE", 6500000, False)
p2.display()

print("=========================")
p3 = Galaxy("S24 Ultra", 1690000, True)
p3.display()

print("=========================")
p4 = Galaxy("Z Flip 5", 1390000, False)
p4.display()

phones = [
	IPhone("IPhone15", 1250000, True),
	IPhone("IPhoneSE", 6500000, False),
	Galaxy("S24 Ultra", 1690000, True),
	Galaxy("Z Flip 5", 1390000, False)
]

for phone in phones:
	print("=========================")
	phone.display()
```
