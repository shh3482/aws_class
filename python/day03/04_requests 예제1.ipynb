```python
# URL에 파라미터 전달 예제
# ?name=홍길동&age=21

import requests

# URL에 접근하여 응답 객체 가져오기
url = "https://search.naver.com/search.naver"
param_dic = {'query': '파이썬'}
response = requests.get(url, params=param_dic)

# URL에 param을 확인
print(response.url)
```

```python
import requests
# 고양이 이미지(구글에서 검색함)
url = "https://cdn.dailyvet.co.kr/wp-content/uploads/2024/05/15231624/20240515ceva_experts3.png"
response = requests.get(url)

# 응답이 성공적(200 OK)이라면 저장
if response.status_code == 200:
    with open("cat.jpg", "wb") as f: 
        f.write(response.content)
        
# URL에 param을 확인
print(response.url)
```

```python
# 큰 파일 다운로드
import requests
# 픽사베이에서 제공하는 무료 영상
url = "https://cdn.pixabay.com/video/2026/03/08/338904_large.mp4"

# stream=True 설정으로 연결만 먼저 열기
with requests.get(url, stream=True) as r:
    r.raise_for_status() # 에러 발생 시 예외 처리
    
    with open("video.mp4", "wb") as f:
        # 8KB씩 조각내서(chunk) 가져오기
        for chunk in r.iter_content(chunk_size=8192):
            f.write(chunk)

print("다운로드 완료!")
```

```python
import requests

url = "https://www.google.com"
response = requests.get(url)

if response.status_code == 200:
    # HTML 소스 코드 출력
    html_data = response.text
    print(html_data[:500])  # 너무 길어서 앞부분 500자만 출력
    
    # 파일로 저장
    with open("index.html", "w", encoding="utf-8") as f:
        f.write(html_data)
```
