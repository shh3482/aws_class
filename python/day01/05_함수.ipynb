```python
nums = [1, 3, 5, 7, 9, 11]

# 리스트에서 가장 큰 수를 반환하는 함수를 선언
def find_max(nums):
    # 초기 최대값
    max_num = nums[0]
    # 반복문으로 리스트에서 최대값을 탐색
    for num in nums:
        # 리스트에서 꺼낸 값이 기존 최대값보다 크면 최대값을 변경
        if num > max_num:
            max_num = num
    # 반복문이 종료 된 후 최대값을 반환
    return max_num

# 함수를 호출해서 가장 큰 수를 조회
print(find_max(nums))
print(max(nums))
```

```python
# match의 활용 - 패턴 매칭
def process_command(command):
    """명령어를 실행하는 함수"""
    print(f"명령어 : {command}")
    match command.split():
        case ["quit"]:
            print("프로그램을 종료합니다.")
        case ["load", file_name]:
            print(f"{file_name} 파일을 불러옵니다.")
        case ["move", x, y]:
            print(f"({x},{y}) 좌표로 이동합니다.")
        case ["print", *contents]:
            print(f"출력 : {contents}")
        case _ :
            print("알 수 없는 명령어 입니다.")

process_command("load data.txt")
print("==================")
process_command("print a b c d")
print("==================")
process_command("move 1 10")
print("==================")
process_command("test")
print("==================")
process_command("quit")
print("==================")
```

```python
# 타입 확인 및 조건부 매칭
def check_value(x):
    match x:
        case int(x) if x > 100:
            print(f"{x}는 정수이고 100이상")
        case int(x):
            print(f"{x}는 정수이고 100이하")
        case str(x):
            print(f"\"{x}\"는 문자열 데이터")
        case _ :
            print(f"{x}는 기타 데이터")
check_value(101)
check_value(100)
check_value("안녕")
check_value(True)
check_value([])
print(f"True를 정수로 변환하면? {int(True)}")
```
