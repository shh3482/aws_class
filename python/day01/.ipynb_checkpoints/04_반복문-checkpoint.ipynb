```python
# for문 : 리스트에서 꺼내오기
과일가게 = ["바나나", "사과", "딸기"]
for 과일 in 과일가게: # 왼쪽 요소가 오른쪽에서 꺼내온거다, 전부 출력해라
    print(f"과일가게에는 {과일}이/가 있습니다.")

# for문 : min부터 max까지 step씩 증가
min, max, step = 1, 10, 3
for i in range(min, max + 1, step):
    print(i)
```

```python
# 리스트 컴프리헨션 예제
# 1 이상 10 이하의 정수를 리스트에 넣어 확인하기
list = [ i for i in range(1, 10 +1) if i % 2 == 0]
print(list)
```

```python
# 1 이상 10 이하의 정수들의 제곱을 리스트에 넣어 확인하기
list = [ i**2 for i in range(1, 10 +1) if i % 2 == 0]
print(list)
```

```python
# 구구단 2단을 while문으로 출력하는 예제
num = 2
i = 1
while i <=9 :
    print(f"{num} X {i} = {num * i}")
    i += 1
```

```python
# num가 소수인지 아닌지 판별하는 코드를 작성하세요
# 단, num는 2이상의 정수라고 가정
num = 4
is_prime = True
for i in range(2, int(num ** 0.5) + 1):
    if num % i == 0:
        is_prime = False
        break
if is_prime:
    print(f"{num}는 소수입니다.")
else:
    print(f"{num}는 소수가 아닙니다.")
```
