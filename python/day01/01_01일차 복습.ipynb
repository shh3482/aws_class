```python
# 정수를 입력받아 입력받은 정수가 홀수인지 짝수인지 출력하는 코드
str = input("정수를 입력하세요 : ")
# 내가 한거
match int(str) % 2 :
    case 0 :
        print(f"{str}은 짝수")
    case _ :
        print(f"{str}은 홀수")

# 파이썬 : 참값 if 조건식 else 거짓값
# print(f"{num}는 {"짝수" if num % 2 == 0 else "홀수"}")
```

```python
# min부터 max까지의 수를 출력하는 함수를 만들고, 1부터 10까지 출력하세요
# 함수명 :  printNums
nums = [ 10, 2, 3, 4, 5, 6, 7, 8, 9, 1 ]
def printNums(min, max) :
    if min > max:
        min, max = max, min
    for i in range(min, max + 1):
        print(i)
printNums(1, 10)
```
