```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
# 데이터 생성

# 1. 데이터 로드
cancer = load_breast_cancer()
```

```python
# 독립, 종속 분리
X = cancer.data
y = cancer.target
```

```python
# 훈련, 테스트 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# 학습 모델을 선정해서 암인지 아닌지를 학습시키고 성능을 출력하세요.
# 학습
# classfier = LogisticRegression()
# classfier.fit(X_train, y_train)
```

```python
# 테스트 데이터 상위 5개를 이용해서 테스트 정보를 출력하고 암 여부를 출력하세요.
# 상위 5개 테스트 데이터 추출

model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)

X_test_selected = X_test # X_test[:5]
y_test_selected = y_test # y_test[:5]
feature_names = cancer.feature_names

for i in range(0,len(X_test_selected)):
  print(f'-------데이터{i+1}-------')
  for j in range(0,len(cancer.feature_names)):
    # print(f'{feature_names[j]} : {X_test_selected[i][j]}')
    pass
  print(f'cancer(predic) : {model.predict([X_test_selected[i]])[0]}')
  print(f'cancer(real) : {y_test_selected[i]}')
  print(f'--------------------')

# print(cancer.feature_names[len(cancer.feature_names)])
```

```python
# 혼동 행렬
from sklearn.metrics import confusion_matrix
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(f'예측(암 X) -> 실제(암 X) : {cm[0][0]}')
print(f'예측(암 O) -> 실제(암 X) : {cm[0][1]}')
print(f'예측(암 X) -> 실제(암 O) : {cm[1][0]}')
print(f'예측(암 O) -> 실제(암 O) : {cm[1][1]}')
```
