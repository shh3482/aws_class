```python
import os
# mnist 이미지를 저장할 폴더를 생성
os.makedirs('mnist_images', exist_ok=True)
```

```python
# mnist 에서 제공하는 숫자 이미지를 다운로드
from sklearn.datasets import fetch_openml

# 데이터 로드. 숫자 이미지를 훈련 6만, 테스트 1만을 제공
mnist = fetch_openml('mnist_784', version=1, as_frame=False)

# 독립, 종속 분리
X, y = mnist['data'], mnist['target']
```

```python
X[0]
```

```python
# 이미지 다운로드 # 꼭 필요한 코드는 아님. 생략 가능함
# 우리가 학습하는 이미지가 어떤 형태의 이미지인지 확인하기 위해 받음
import numpy as np
import cv2
import os  # os 모듈 임포트 확인!

# [수정 포인트] 1000 대신 len(X)를 사용하여 실제 데이터 개수만큼만 반복합니다.
for i in range(len(X)):
    # X[i]가 1차원(784)이므로 (28, 28)로 변환
    image = X[i].reshape(28, 28).astype(np.uint8)
    label = y[i]
    
    # 폴더 생성 (label이 문자열인지 확인)
    os.makedirs(f'mnist_images/{label}', exist_ok=True)
    
    # 이미지 파일 저장
    cv2.imwrite(f'mnist_images/{label}/img_{i}.png', image)

print(f"총 {len(X)}개의 이미지 다운로드 완료!")
```

```python
# 이미지를 학습 시키기 위한 데이터로 변환
# X : 1차 행렬 리스트
# y : 1차 행렬 리스트의 답
X_full, y_full = X, y

import glob
import cv2
import numpy as np

X, y = [], []

# 폴더별로 이미지들을 가져와서 X, y에 넣어주는 작업
for folder in range(10):
    
    # 폴더에 있는 이미지(.png)들을 가져오는 경로 설정
    file_path = f'mnist_images/{folder}/*.png'
    
    # 폴더에 있는 이미지들을 하나씩 꺼내서 작업
    for file in glob.glob(file_path):
        img = cv2.imread(file, cv2.IMREAD_GRAYSCALE)
        
        if img is not None: # 이미지를 정상적으로 읽었는지 확인
            X.append(img.flatten())
            y.append(str(folder))


# [수정 포인트] 모든 반복문이 완전히 끝난 뒤에 리스트를 numpy 배열로 변환합니다.
X = np.array(X)
y = np.array(y)

# 결과 확인
print(X.shape, y.shape)
```

```python
# 훈련, 테스트 데이터 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.svm import SVC

model = SVC(kernel='rbf', C=1.0)
model.fit(X_train,y_train)
```

```python
# 200개 데이터로 정확도 확인
model.score(X_test, y_test)
```

```python
# 7만개 데이터를 훈련, 테스트로 분리
X_full_train, X_full_test, y_full_train, y_full_test = train_test_split(X_full, y_full, test_size=0.2)

# 학습
model2 = SVC(kernel='rbf', C=1.0)
model2.fit(X_full_train,y_full_train)
```

```python
model2.score(X_full_test, y_full_test)
```

```python
model2.score(X_test, y_test)
```

```python
# 이미지 불러오기
test_img = cv2.imread('test_img.png', cv2.IMREAD_GRAYSCALE)

# 이미지 형태를 변환 => 1차 행렬
test_img = cv2.resize(test_img, (28, 28))
test_data = test_img.flatten().reshape(1,-1)
```
