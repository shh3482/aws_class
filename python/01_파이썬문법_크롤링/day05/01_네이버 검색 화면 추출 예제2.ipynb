```python
# naver_nl_ai_페이지.csv 파일에 있는 뉴스 링크로 가서 기사 내용을 가져와서,
# naver_nl_ai_페이지_detail.csv를 생성하는 예제
# naver_nl_ai_페이지_detail.csv 열 이름 => title, href, detail

import requests
import pandas as pd
from bs4 import BeautifulSoup

# csv 파일 읽기
df = pd.read_csv('naver_news_ai.csv', encoding='utf-8')

# 기사 내용 가져오기

# 기사 내용을 담을 테이블 생성
details = []

for link in df['href']:
  
  # df테이블에 링크로 접속
  response = requests.get(link)
  
  # 텍스트 변환
  soup = BeautifulSoup(response.text, 'lxml')
  
  # 기사 내용 부분 찾아서 content에 저장
  content = soup.select_one('#dic_area')
  
  # 기사 내용에 content 를 저장
  detail = content.text.strip()
  
  # 기사 내용들에 기사 내용을 추가
  details.append(detail)

# 기사 내용들 테이블에 방금 만든 기사 내용들을 담기
df['detail'] = details
print(df)
# 기사 내용을 naver_nl_ai_페이지_detail.csv 로 저장
df.to_csv('naver_nl_ai_page_detail.csv', index=False, encoding='utf-8-sig')
  

# naver_nl_ai_페이지_detail.csv에 저장된거를 열 이름 => title, href, detail 로 정리
```

```python
# WBC 뉴스 20개를 가져와서  csv파일 1개로 저장
# title, href, article

import requests
import pandas as pd
from bs4 import BeautifulSoup
import time

keyword = input("키워드를 입력하세요 : ")

print(f"입력하신 키워드는 : {keyword} 입니다.")

def search(keyword, start:int=1):

  titles = []
  hrefs = []
  i = 1

  while len(titles) < 20:

    url = 'https://search.naver.com/search.naver'

    params = {
        'where': 'news',
        'query': keyword,
        'start': i
    }

    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)',
        'Referer':'https://www.naver.com/'
    }

    response = requests.get(url, params=params, headers=headers)

    if response.status_code == 200:
      print(f"{i} 페이지 접속 성공")

      soup = BeautifulSoup(response.text,'lxml')

      links_divs = soup.select('.n8_DDlCjHxLYRO50CY50')

      for link_div in links_divs:

        title_link = link_div.select_one('.XEtVZ4N7DI2Pv29xe7f3')
        href_link = link_div.select_one('.RQNKk8QZaQZble3gmgEj [data-heatmap-target=".nav"]')

        if title_link and href_link:
          titles.append(title_link.text)
          hrefs.append(href_link.get("href"))
          
      i += 10

    time.sleep(2)

  data = {
    'title': titles[:20],
    'href': hrefs[:20]
  }

  df = pd.DataFrame(data)
  
  return df

df = search(keyword)

details = []
count = 1

print(f"csv파일을 생성하여 기사들을 저장하겠습니다.")

for link in df['href']:
  
  response = requests.get(link)
  
  soup = BeautifulSoup(response.text, 'lxml')
  
  content = soup.select_one('#dic_area')
  
  if content:
    detail = content.text.strip()
  else:
    detail = ""
  
  details.append(detail)
  
  
  print(f"{count}/20 csv에 기사 저장완료")
  count += 1

  time.sleep(1)

df['article'] = details

df.to_csv(f'naver_nl_{keyword}.csv', index=False, encoding='utf-8-sig')

print(df)



```
