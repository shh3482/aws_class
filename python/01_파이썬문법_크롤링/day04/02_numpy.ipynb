```python
import numpy as np

# 홍길동 학생의 국어, 영어, 수학, 성적을 입력받아 numpy 배열로 만들고 콘솔에 출력하세요
# 콘솔에서 입력받을 때 사용할 문구=> 성적을 입력하세요 : 

# 성적을 담을 빈 배열 생성
scores = []

# 콘솔에서 입력받기
scores = input("성적을 입력하세요 : ")

# 공백 기준으로 나누고 문자열을 정수로 변환해서 배열에 저장
scores = list(map(int, scores.split()))

arr = np.array(scores)

# 콘솔에서 입력 받은거 배열에 넣기
print(f"국어 성적: {arr[0]}점, 영어 성적: {arr[1]}점, 수학 성적: {arr[2]}점")
```

```python
# 학생의 평균을 출력하세요
print(f"전체 평균 : {arr.mean()}")
```
