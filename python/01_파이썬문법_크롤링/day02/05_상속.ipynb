```python
# 상속 예제
# 자동차 class
class Car:
    def __init__(self, color : str, type :  str, name : str):
        self.__color = color
        self.__type = type
        self.__name = name

    def print(self):
        print(f"차종 : {self.__type}")
        print(f"차명 : {self.__name}")
        print(f"색상 : {self.__color}")

    def powerOn(self):
        print("시동을 켭니다.")
        
    def powerOff(self):
        print("시동을 끕니다.")

class Hcar(Car):
    company : str = "현대"

    def __init__(self, color : str, type :  str, name : str):
        super().__init__(color, type, name)

    def print(self):
        print("-------------------")
        print(f"회사 : {Hcar.company}")
        super().print()
        print("-------------------")
    
c1 = Car("검정", "소형", "아반떼")
c1.print()
c2 = Hcar("파랑", "소형", "아반떼")
c2.print()
```

```python
# 다중 상속인 경우, 매서드 명이 겹치면 왼쪽에 있는 클래스가 적용
class P1: 
    def print(self):
        print("P1 클래스입니다.")

class P2: 
    def print(self):
        print("P2 클래스입니다.")

class C(P1,P2):
    pass
c1 = C()
c1.print()
```
