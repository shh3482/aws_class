```python
# 산술연산자 : /
print( f"{1} / {2} = {1/2}")

# 산술연산자 : // => 몫을 추출
print( f"{1} // {2} = {1//2}")

# 산술연산자 : ** => 제곱
print( f"{2}의 {3}제곱 = {2 ** 3}")

# 산술연산자 : * => 문자열 * 정수 => 정수만큼 반복
print( f"{"hi"} * {3} = {"hi" * 3}")
```

```python
# 논리 연산자
# && => and
score = 95
print(f"{score}점은 A학점인가요? {score >= 90 and score <= 100}")

# || => or
민증 = True
운전면허증 = False
print(f"신분증을 가지고 있나요? {민증 or 운전면허증}")

# ! => not
print(f"신분증이 없나요? { not (민증 or 운전면허증)}")
```

```python
# 멤버십 연산자 : in, not in
과일가게 = ["바나나", "사과", "포도", "딸기"]
print(f"과일 가게에서 파는 과일 목록 : {과일가게}")

print(f"가게에 수박이 있나요? {"수박" in 과일가게 }")
print(f"가게에 수박이 없나요? {"수박" not in 과일가게 }")
```
